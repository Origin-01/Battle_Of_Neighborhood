```python
# Utilizing the Foursquare API to explore the neighborhood and Segment Them

# Define Foursquare Credentials and Version
CLIENT_ID='2YRUFLK1XMVKN1015KLEIN0LPJ4HPZQDEZKSVSSFVKRYNJUW'
CLIENT_SECRET='XJ3ZNUTX0VHNMTCQBEF0JYLSRJTDN5JOMHMN4SMXRZSH1GWB'
VERSION='20180605'
limit=100

print("Client id {}, Client Secret {}, Version {}".format(CLIENT_ID,CLIENT_SECRET,VERSION))
```

```python
# Explore Neighborhoods in Toronto

# Let's create a function to repeat the process to all the neighborhoods in Toronto
def getNearbyVenues(names, latitudes,longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names,latitudes,longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            limit)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Run the above function
toronto_venues= getNearbyVenues(names=toronto_new['Neighborhood'],latitudes=toronto_new['Latitude'],longitudes=toronto_new['Longitude'])
```

```python
# Unique categories curated from all the returned venues
len(toronto_venues['Venue Category'].unique())
```

```python
# ANALYZE EACH NEIGHBORHOOD

# One Hot Encoding
toronto_oneshot=pd.get_dummies(toronto_venues[['Venue Category']],prefix="",prefix_sep="")

# Add Neighborhood Column Back to dataframe
toronto_oneshot['Neighborhood']=toronto_venues['Neighborhood']

# Moving the Neighborhood column to the front
col=[toronto_oneshot.columns[-1]]+list(toronto_oneshot.columns[:-1])
toronto_oneshot=toronto_oneshot.reindex(columns=col)

# Grouping Rows by neighborhood by taking the mean of the frequency of occurence of each category
toronto_grouped=toronto_oneshot.groupby('Neighborhood').mean().reset_index()

toronto_grouped.head()
```

```python
# Now to sort out the dataframe lets print each neighborhood along with their top 5 most common venue

numtop_venues=5

for name in toronto_grouped['Neighborhood']:
    print('-----'+name+'-----')
    
    temp=toronto_grouped[toronto_grouped['Neighborhood']==name].T.reset_index()
    temp.columns=['Venue','frequency']
    temp=temp.iloc[1:]
    temp['frequency']=temp['frequency'].astype(float)# Converting the string to float
    temp=temp.round({'frequency':2})
    
    print(temp.sort_values('frequency',ascending=False).reset_index(drop=True).head(numtop_venues))
    print('\n')
```

```python
# A function to sort the venues in descending order
def return_most_common_venues(row,num):
    row_cat=row.iloc[1:]
    row_cat_sort=row_cat.sort_values(ascending=False)
    
    return row_cat_sort.index.values[0:num]
```

```python
# Lets put the above data in pandas dataframe as the top 10 venues of each Neighborhood

num=10
indicators=['st','nd','rd']

# create columns according to number of top 10 venues
columns=['Neighborhood']

for i in np.arange(num):
    
    try:
        columns.append('{}{} Most Common Venue'.format(i+1,indicators[i]))
    except:
        columns.append('{}th Most Common Venue'.format(i+1))
        
# Create the dataframe
venue_sorted=pd.DataFrame(columns=columns)
venue_sorted['Neighborhood']=toronto_grouped['Neighborhood']

for i in np.arange(toronto_grouped.shape[0]):
    venue_sorted.iloc[i,1:]=return_most_common_venues(toronto_grouped.iloc[i, :],num)

venue_sorted
```

Step 5 - Clustering of data (K-Means)
---

```python
# Set Number of KClusters
kcluster=5

toronto_grouped_clust=toronto_grouped.drop('Neighborhood',1)

# Run k-means clustering
kmeans=KMeans(n_clusters=kcluster,random_state=0).fit(toronto_grouped_clust)

# checking the cluster labels generated for each row in the dataframe
kmeans.labels_[0:20]
```

```python
# add clustering labels
venue_sorted.insert(0,'ClusterLabel',kmeans.labels_)

toronto_merged=toronto_new

#merge toronto_grouped with toronto_data to add lat/long for each neighborhood
toronto_merged=toronto_merged.join(venue_sorted.set_index('Neighborhood'),on='Neighborhood')



toronto_merged.head(20)
```

```python
toronto_merged.dropna(inplace=True)
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kcluster)
ys = [i+x+(i*x)**2 for i in range(kcluster)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['ClusterLabel']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster-1)],
        fill=True,
        fill_color=rainbow[int(cluster-1)],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
